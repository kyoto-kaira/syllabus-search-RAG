```python
def or_search(input,key):
  a = set()
  for i,metadata in enumerate(metadatas):
    if key in metadata.keys():
      if not metadata[key] in input:
        a.add(str(i))
    else:
      a.add(str(i))
  return a
```

```python
def or_list_search(input,data):
  a = set()
  for i,y in enumerate(data):
    b = 0
    for d in y:
      if  d in  input:
        b = 1
    if b == 0:
      a.add(str(i))
  return a
```

```python
def word_search(input,texts):
  a = set()
  for i,text in enumerate(texts):
    if not input in text:
      a.add(str(i))
  return a
```

```python
def hyoka_search(input,ratio):
  seiseki = []
  for d in hyoka:
    a = set()
    for k,v in d.items():
      if isinstance(v, dict):
        for assessment_type, score in v.items():
          if score >= ratio:
            a.add(assessment_type)
      elif v >= ratio:
        a.add(k)
    if len(a) == 0:
      a.add('なし')
    seiseki.append(a)
  return or_list_search(input,seiseki)
```

```python
def ids_union(ids):
  if len(ids) == 1:
    return list(ids[0])
  else:
    for i in range(len(ids)-1):
      if i == 0:
        a = ids[0] | ids[1]
      else:
        a = a | ids[i+1]
    return list(a)
```
